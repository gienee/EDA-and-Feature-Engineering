# Handle OHE in case of multiple unique values in a categorical features

```python
df=pd.read_csv('mercedesbenz.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
df.head()
```

```python
###Let;s take the example of X1 where amy unique features are present so,we donot apply OHE on all the unique features
```

```python
df['X1'].unique()
```

```python
len(df['X1'].unique())     ## Total no. of unique values is 27
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10)     ## Take out top 10 frequent unique value and apply OHE on only those unique values
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)    
lst_10
```

```python
import numpy as np

for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)     ## Apply OHE on only top 10 most frequent values which are present
                                                                ## in any categorical columns
lst_10.append('X1')
df[lst_10]
```

# Handling the categorical features by the Count Or Frequency Encoding

Here we convert the categorical value into numeric value by replacing the categorical value by the number of frequency it comes...so,first we find the frequency or count of each categorical value and then replace that word with that frequency

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
train_set=train_set[columns]

train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']    ##Gives the name  of each column
train_set.head()
```

```python
##Let's take the example of Country and convert it into numeric column

train_set["Country"].unique()
```

```python
len(train_set["Country"].unique())
```

```python
train_set["Country"].value_counts()
```

```python
## Observation: Here we have to replace the ? with NAN values and then apply the most frequent method to handle the missing value NAN with that frequent word
#And then handle the categorical value
```

```python
country_map=train_set['Country'].value_counts().to_dict()
                                          ## it will not create any new columnit will do the mapping only in that existed column
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

# Advantages

1. Easy To Use
2. Not increasing feature space

# Disadvantages

1. It will provide same weight if the frequencies are same

# Handling the Categorical columns by Target Guided Ordinal Encoding

1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
df=pd.read_csv('Titanic-Dataset.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)       ##Handle the missing value by the random imputation i.e the word Missing
```

```python
df.head()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]                    ##Take out the first letter of the column "Cabin" 
df.head()
```

```python
df['Cabin'].unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}          ##enumerate means iterate over the column
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)                  ##It is a type of ordinal encoding i.e mapping
df.head()
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
```
